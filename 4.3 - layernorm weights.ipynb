```python
import transformers as trf
import torch as pt
import numpy as np
from collections import Counter, OrderedDict
import matplotlib.pyplot as plt
```

```python
bert = trf.BertForMaskedLM.from_pretrained('bert-base-uncased')
```

```python
with pt.no_grad():
    weight = bert.cls.predictions.transform.LayerNorm.weight.detach()
    bias = bert.cls.predictions.transform.LayerNorm.bias.detach()
```

```python
fig = plt.figure(dpi=120)
ax = fig.subplots()
ax.plot(weight)
ax.set_xlabel('dimension of layernorm weights')
ax.set_ylabel('layernorm weight value')
fig.savefig('layernorm-weights.png')
plt.show()
plt.close(fig)
```

```python
fig = plt.figure(dpi=120)
ax = fig.subplots()
ax.plot(bias)
ax.set_xlabel('dimension of layernorm bias')
ax.set_ylabel('layernorm bias value')
fig.savefig('layernorm-bias.png')
plt.show()
plt.close(fig)
```
